```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/Cleaned_Superstore.csv")
```

```python
sales_by_category = df.groupby(
    "Category"
)["Sales"].sum()
```

```python
plt.bar(
    sales_by_category.index,    # This gives category name
    sales_by_category.values,     # This gives sales values
)
plt.title('Total sales by category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.savefig("../images/sales_by_category.png")   # use to save chart in image folder
plt.show()
```

```python
profit_by_category = df.groupby(
    "Category"
)["Profit"].sum()
```

```python
plt.bar(
    profit_by_category.index,
    profit_by_category.values
)
plt.title("Total profit by category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.savefig("../images/profit_by_category.png")
plt.show()
```

```python
sales_by_region = df.groupby(
    "Region"
)["Sales"].sum()
```

```python
plt.bar(
    sales_by_region.index,
    sales_by_region.values
)
plt.title("Total sales by region")
plt.xlabel("Region")
plt.ylabel("Total sales")
plt.savefig("../images/sales_by_region.png")
plt.show()
```

```python
sales_by_subcategory = df.groupby(
    "Sub-Category"
)["Sales"].sum().sort_values(
    ascending = False
)
```

```python
# because of there is so many subcategories so that here we use horizontal chart
plt.barh(
    sales_by_subcategory.index,
    sales_by_subcategory.values
)
plt.title("Sales by sub-category")
plt.xlabel("Total sales")
plt.ylabel("Sub-Category")
plt.savefig("../images/sales_by_subcategory.png")

plt.show()
```

```python
profit_by_subcategory = df.groupby(
    "Sub-Category"
)["Profit"].sum().sort_values(
    ascending = False
)
```

```python
plt.barh(
    profit_by_subcategory.index,
    profit_by_subcategory.values
)
plt.title("Profit by sub-category")
plt.xlabel("Total Profit")
plt.ylabel("Sub-Category")
plt.savefig("../images/profit_by_subcategory.png")

plt.show()
```
