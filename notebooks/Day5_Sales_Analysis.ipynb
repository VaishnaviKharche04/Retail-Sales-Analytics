```python
import pandas as pd
df = pd.read_csv("../data/Cleaned_Superstore.csv")
```

```python
df.head()
```

```python
# which category generates the highest total sales?
sales_by_category = df.groupby("Category")["Sales"].sum()
sales_by_category
```

```python
# which category generates the highest total profit?
sales_by_category = df.groupby("Category")["Profit"].sum()
sales_by_category
```

```python
# which region generates the highest sales?
sales_by_region = df.groupby("Region")["Sales"].sum()
sales_by_region
```

```python
# which region generates the highest profit?
sales_by_region = df.groupby("Region")["Profit"].sum()
sales_by_region
```

```python
sales_by_segment = df.groupby("Segment")["Sales"].sum()     # sales by segment
```

```python
sales_by_segment
```

```python
category_analysis = df.groupby("Category").agg(
    Total_sales = ('Sales','sum'),
    Total_profit = ('Profit','sum'),
    Average_sales = ('Sales','mean'),
    Average_profit = ('Profit','mean')
)

category_analysis
```

```python
# Find the best category

category_analysis.sort_values(
    by="Total_profit",
    ascending=False
)
```
